# Напишите функцию, которая принимает строку и выводит ее в обратном порядке.

Пример: СТРОКА

Вывод: АКОРТС 

```python
#получаем ввод от пользователя и присваиваем значение переменной input_string
input_string = input(f"Введите строку,которую требуется перевернуть\n")

#создаём функцию, которая получает строку на вход и переставляет символы в обратном порядке
def reverse_string(input_string):
    return input_string[::-1]

#вызываем функцию и присваиваем результат выполнения переменной reversed_string
reversed_string = reverse_string(input_string)

#выводим результат 
print(reversed_string)
```

Напишите функцию, которая принимает два аргумента: строка и символ. Функция должна вернуть количество появлений символа в строке.
Пример: 

яблоко

о

2


```python
#присваиваем переменным input_string и char, значения строки и символа соответственно
input_string = "Hello, world!"
char = "o"

#создаём функцию, которая ищет в заданной строке (input_string) количество вхождения указанного символа (char)
def count_char(input_string, char):
    count = 0
    for c in input_string:
        if c == char:
            count += 1
    return count

#вызываем функцию и присваиваем результат переменной char_count
char_count = count_char(input_string, char)

#выводим результат подсчета количества символов в строке
print(f"The character '{char}' appears {char_count} times in the string '{input_string}'")
```

решение с вводом условий от пользователя

```python
#создаём функцию, которая ищет в заданной строке (str_input) количество вхождения указанного символа (char)
def count_char(str_input, char):
  count = 0
  for i in str_input:
    if i == char:
      count += 1
  return count

#получаем ввод строки и символа от пользователя
str_input = str(input('Введите фразу, в которой требуется найти символ:\n'))
char = str(input('Введите символ, который требуется найти:\n'))

#вызываем функцию и присваиваем результат переменной char_count
char_count = count_char(str_input, char)

#выводим результат подсчета количества символов в строке
print(f"Всего символов '{char}' во фразе '{str_input}' найдено '{char_count}' ")
```

Напишите функцию, которая принимает список строк и возвращает самую длинную строку из списка.

Пример:
яблоко, клубника, ананас

Ответ:
клубника


```python
#задаём массив слов и присваиваем переменной
list1 = ["банан","клубника","ананас"]

#создаём функцию, которая вычисляет самое длинное слово и определяет количество символов в нем
def list_len_count(new_list):
  max_len = 0
  for each in new_list:
    if len(each) > max_len:
      max_len = len(each)
      word = each
  return(max_len,word)


#вызываем функцию и присваиваем результат переменной result, т.к. функция возвращает два значения, то в переменную result записывается массив из двух значений
result = list_len_count(list1)

#выводим ответ
print(f"В самом длинном слове '{result[1]}' найдено {result[0]} символов")
```

Напишите функцию, которая принимает список чисел и проверяет, есть ли в списке повторяющиеся числа.

Пример:
2 2 4 2 

Вывод:
Есть

Пример:
2 3 1 4

Вывод:
нет

```python
#задаём массив чисел и присваиваем переменной num_list
num_list = [1,2,3,4,5,6]

#создаём функцию для определения наличия повторяющихся чисел
def check_num_list(num_list):
  i = 0
  len_n = len(num_list)
  for each in num_list:
    for n in range(i+1, len_n):
      if each == num_list[n]:
        return(True)
    i = i + 1
  return(False)

#вызываем функцию и выводим результат 
print(check_num_list(num_list))
```
