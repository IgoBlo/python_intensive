```python
#  student,   name, age, score
class student():
  def __init__(self,name,age,score):
    self.name = name
    self.age = age
    self.score = score

#  avr_score   student,      
  def avr_score(self):
    sum = 0
    count = 0
    for each in self.score:
      sum += each
      count += 1
    sum = sum/count
    print(f':    {self.name}\n. : {sum}')

#  student1  student
student1 = student("",4,[5,2,4,5,4])

#  avr_score      student1
student1.avr_score()
```

```python
#  worker,   name, salary, position
class worker():
    def __init__(self,name,salary,position):
      self.name = name
      self.salary = salary 
      self.position = position

#  award   worker,       , (  * )    . 
    def award(self,p_award):
      award = self.salary * p_award
      print(f':  {self.name}\n:     {award}')

#  worker1  worker
worker1 = worker("",12000,"director")

#   
p_award = 1.5

#  award      worker1
worker1.award(p_award)
```

```python
#  figure,     "sides" (  )
class figure():
  def __init__(self,sides):
    self.sides = sides

#   
  def perimeter(self):
    sum = 0
    for each in self.sides:
      sum += each
    print(f': {sum}')

#  fig  figure    
fig = figure([4,5,6,6,5])

#  perimetr  fig
fig.perimeter()
```
