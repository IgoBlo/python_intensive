Напишите класс «Студент», который имеет атрибуты «имя», «возраст» и «оценки». Оценки должны быть списком. Добавьте метод «средняя_оценка», который вычисляет среднюю оценку студента.

В качестве методолгии вычисления среднего возьмем "среднее арифметическое", т.е. формула для расчета будет: *сумма оценок / количество оценок*.

```python
#создаем класс student, с атрибутами name, age, score
class student():
  def __init__(self,name,age,score):
    self.name = name
    self.age = age
    self.score = score

#создаем метод avr_score для класса student, который позволяет вычислять среднюю оценку студента
  def avr_score(self):
    sum = 0
    count = 0
    for each in self.score:
      sum += each
      count += 1
    sum = sum/count
    print(f'Студент:    {self.name}\nСр. оценка: {sum}')

#создаём объект student1 класса student
student1 = student("Ваня",4,[5,2,4,5,4])

#вызываем метод avr_score для расчета средней оценки объекта student1
student1.avr_score()
```

Напишите класс «Работник», который имеет атрибуты «имя», «зарплата» и «должность». Добавьте метод «размер_премии», который принимает процент премии и вычисляет размер премии в зависимости от зарплаты работника.

```python
#создаем класс worker, с атрибутами name, salary, position
class worker():
    def __init__(self,name,salary,position):
      self.name = name
      self.salary = salary 
      self.position = position

#создаем метод award для класса worker, который принимает на вход значение процента премии, вычисляет(процент премии * зарплата) и выводит значение премии. 
    def award(self,p_award):
      award = self.salary * p_award
      print(f'Сотрудник:  {self.name}\nпремия:     {award}')

#создаём объект worker1 класса worker
worker1 = worker("Денис",12000,"director")

#задаем процент премии 
p_award = 1.5

#вызываем метод award для расчета премии для объекта worker1
worker1.award(p_award)
```

Напишите класс «Фигура», который имеет атрибут «стороны». Стороны должны быть списком. Добавьте метод «периметр», который вычисляет периметр фигуры.

Доработать ввод с клавиатуры

```python
#создаем класс figure, для которого указывается атрибут "sides" (массив сторон фигуры)
class figure():
  def __init__(self,sides):
    self.sides = sides

#указываем метод вычисления периметра
  def perimeter(self):
    sum = 0
    for each in self.sides:
      sum += each
    print(f'Периметр: {sum}')

#создаём объект fig класса figure и присваиваем значения сторон
fig = figure([4,5,6,6,5])

#вызываем метод perimetr объекта fig
fig.perimeter()
```

Напишите класс для героя в фэнтезийном мире. Класс должен включать  имя героя, здоровье, показатель силы. Напишите метод, который будет принимать СИЛУ противника и УРОН и вычитать УРОН из здоровья героя, если СИЛА противника больше чем СИЛА героя

```python
#задаем значения переменных
enemy_strong = 10
enemy_damage = 10
name = 'Геройское имя'
health = 100
strong = 9

#создаем класс hero
class hero():
  def __init__(self,name,health,strong):
    self.name = name
    self.health = health
    self.strong = strong

#создаем метод для получения урона героем
  def get_kick(self):
    if self.strong < enemy_strong:
      self.health -= enemy_damage
      print(f'Герой "{self.name}" получил урон. Текущее здоровье: {self.health}')
    elif self.strong > enemy_strong:
      print(f'Вы не получили урон, ваша сила больше вражеской')
    else:
      print(f'Ваши силы равны')

#создаем объект класса hero
hero1 = hero(name,health,strong)

#вызываем метод получения урона
hero1.get_kick()
hero1.get_kick()
hero1.get_kick()
```
