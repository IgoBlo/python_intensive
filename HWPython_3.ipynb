Задача 1. Мастер-джедай дал юному падавану список чисел и попросил к закату принести сумму, не пользуясь функцией sum()
Помогите падавану реализовать программу


```python
#присваиваем переменной массив значений, которые хотим просуммировать
number = [1,24,400,303,23,33,44,4]

#объявляем переменную для сохранения суммы
sum = int()

#с помощью цикла for перебираем и суммируем каждый элемент массива number
for each in number:
  sum = sum + each

#выводим результат
print(sum)
```
